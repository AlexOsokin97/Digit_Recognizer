```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
data = pd.read_csv('train.csv')
```

```python
'''
getting the row numbers where we have a label = 0
and choosing 9 different pixel sets to get a general view of how the label looks like
'''
zeros = data.loc[data['label'] == 0]
zeros.head(9)
```

```python
zero1 = np.array(zeros.iloc[1, 1:], dtype=np.uint8)
```

```python
image = Image.fromarray(zero1)
image.save('new.png')
```
