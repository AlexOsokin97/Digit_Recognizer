```python
'''
Checking how many samples of each digit label (0-9) are in the training set to see 
if there is a big imbalance between the labels which could affect the learning
algorithm's performance
'''
num_digits = {} #creating an empty dictionary
i = 0 #creating a loop variable which will correspond with the 'label' column values by taking up values from 0-9

while i <= 9:
    num_digits[i] = len(data.loc[data['label'] == i])
    i+=1

num_digits
```

```python
import seaborn as sns

sns.countplot(x = data['label'], data=data, palette='Set1')
```
