```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
data = pd.read_csv('train.csv')
```

```python
'''
getting the row numbers where we have a label = 0
and choosing 9 different pixel sets to get a general view of how the label looks like
'''
zeros = data.loc[data['label'] == 0].copy()
zeros.drop(['label'], axis=1, inplace=True)
zeros.head(9)
```

### in order to form an image from the given pixels I firstly do a Normalization in order to reduce the effect of lighting differences. This can be achieved by dividing each value by 255.0, which scales the values between 0-1. Secondly I'll reshape the image into 3 dimensions (height=28px, length = 28px, canal = 1)

```python
zeros = zeros / 255.0
zeros = zeros.values.reshape(-1, 28, 28, 1)
```

```python
import random

plt.figure(figsize=(12,6))
for i in range(12):
    g = random.randint(0,30)
    plt.subplot(3, 4, i + 1)
    plt.imshow(zeros[g][:,:,0], cmap='gray')
    plt.axis('off') 
    
plt.suptitle('Sample Images of The Digit 0', fontsize=14, y=1.05)
plt.tight_layout()   
```

### Because the method is the same for each digit I'll create a function that prints and saves each digit grid automatically. This will save me time and alot of lines of code

```python
def show_digit(name, digit):
    '''
    a function that displays 12 gray scaled images in a (3,4) shape.
    
    Input:
            name (string): the grid name for file saving. (example: 'somename.(jpeg/pdf/png)')
            digit (integer): The Digit for image generation (0-9)
    output:
            an image grid file which is saved on your local directory
    example:
            import random
            
            var = data.loc[data['label'] == digit(0)].copy()
            var.drop(['label'], axis=1, inplace=True)

            var = var / 255.0
            var = var.values.reshape(-1, 28, 28, 1)

            plt.figure(figsize=(12,6))
            for i in range(12):
                g = random.randint(0,30)
                plt.subplot(3, 4, i + 1)
                plt.imshow(var[g][:,:,0], cmap='gray')
                plt.axis('off') 
    
            plt.suptitle('Sample Images of The Digit '+str(digit)(0), fontsize=14, y=1.05)
            plt.tight_layout()   
            plt.savefig(name('0_grid.jpeg'))
    '''
    
    var = data.loc[data['label'] == digit].copy()
    var.drop(['label'], axis=1, inplace=True)

    var = var / 255.0
    var = var.values.reshape(-1, 28, 28, 1)

    plt.figure(figsize=(12,6))
    for i in range(12):
        g = random.randint(0,30)
        plt.subplot(3, 4, i + 1)
        plt.imshow(var[g][:,:,0], cmap='gray')
        plt.axis('off') 
    
    plt.suptitle('Sample Images of The Digit '+str(digit), fontsize=14, y=1.05)
    plt.tight_layout()   
    plt.savefig(name)
```

```python
for i in range(0,10):
    name = str(i)+'_grid.jpeg'
    show_digit(name, i)
```
